Ejercicio 1: Crear una clase llamada Persona con los atributos: nombre, edad y
DNI. Diseñar los siguientes métodos para la clase:

● Un constructor, donde los datos pueden estar vacíos;

● Mostrar los datos de las personas;

● Determinar si es mayor de edad.


```python
class Persona:
  def __init__(self,nom=None,eda=None,dni=None):
    self.nombre=nom
    self.edad=eda
    self.dni=dni

  def mostrar_datos(self):
    print("{"+f"\nNombre: {self.nombre}")
    print(f"Edad: {self.edad}")
    print(f"Dni: {self.dni}\n"+"}")

  def es_mayor(self):
    if self.edad==None:
      return False
    elif self.edad>17:
      return True
    else:
      return False

persona1=Persona()
persona2=Persona("Carlos")
persona3=Persona("Carmen",15,12345678)
persona4=Persona("Ramiro",19,12345679)



persona1.mostrar_datos()
persona2.mostrar_datos()
persona3.mostrar_datos()
persona4.mostrar_datos()
tupla=(persona1,persona2,persona3,persona4)
for i in tupla:
  print(f"{i.nombre} con DNI N°")
  if i.es_mayor():
    print("Es mayor")
  elif (not i.es_mayor()) and i.edad!=None :
    print("Es menor")
  else: print("No tiene edad asignada")
```
